```python
result = []
nums = [1, 2, 3, 4, 5]
for x in nums:
    if x > 3:
        if x % 2 == 0:
            result.append(f'{x}는 3보다 큰 짝수')
        else:
            result.append(f'{x}는 3보다 큰 홀수')
print(result)
```

```python
def sumthingFunction(i): # 함수 sumthingFunction(i)을 생성
    if i % 100 ==0:      # 조건문 100으로 나눴을때 나머지가 0이 되는것
        return i         # 조건에 일치하면 i를 리턴 일치하지 않으면 0을 리턴하는데,
    else:
        return 0

l = [100, 200, 300, 101, 202, 303]
l = [sumthingFunction(i) for i in l] # l(리스트)에 있는 값들을 하나씩 꺼내서, 함수에 태워보고, 조건에 맞는것만 출력
l # 출력: [100, 200, 300, 0, 0, 0]
print(l)
```

```python
# 우리가 맞춰야할 숫자!
import random

secret_number = random.randrange(1, 101)
```

```python
# 정답을 위한 Step 1
# if문을 사용해서 같으면 정답 틀리면 틀렸습니다. 라고 출력해보기!
if number == secret_number:
    print("정답")
else:
    print("틀렸습니다.")
```

```python
# step 1을 구현했는데...
# 최대 100번까지 맞춰야한다.
# input을 계속 입력해야한다.
# 힌트가 없다.

# Step 2 -> 20번을 반복하는 기능을 구현! for문
```

```python
# Step 3. 왜 틀렸는데?! Up/down 힌트를 주면 좋을것 같다.

#for문을 추가하면서 재배치!

secret_number = random.randrange(1, 101) # 고정값! # Step 3. secret_number은 고정 값

for i in range(20): # Step1. 반복할 값을 지정한다.
    print(f'{i+1} 번째 숫자를 입력하세요(1에서 100까지)')
    number = int(input('숫자를 입력하세요~ (1에서 100사이)')) # Step 2. input을 지정한다.
    if number == secret_number: # Step 4 : 조건문
        print("정답") # 어떻게 끝내는거지?
        # break 빠져나가게 하는 구문!
    elif number > secret_number:
        print("틀렸는데, 값이 더 작아요.")
    else: # 틀리긴 틀렸는데 조건이 세부적으로 되었으면 좋겠다.!! elif / else
        print("틀렸는데, 값이 더 커요")

# 계속 틀리다가 맞추면 정답! 출력 하긴한다. 20번을 무조건 반복한다.
# up & Down 최대한 if / elif / else
```
